```
import pandas as pd
```

```
#Load the full Portfolio Catalogue
pfl = pd.read_csv('PortfolioUsage_March152023.csv')
```

```
pfl.shape
```

```
pfl['description_len'] = pfl['NVS_Course_Description'].str.len()
# pfl.pivot_table(values='description_len', index='NVS_Course_Training_Type', columns='NVS_Course_Training_Provider').fillna(0)
```

```
pfl_subset = pfl.query("""
    NVS_Course_Training_Provider.isin(['Coursera', 'Coursera - old', 'Get Abstract', 'LinkedIn Learning', 'TED Talks', 'Other', 'External Training', 'Adobe Connect'])==False
    """)
print(pfl_subset.shape)
```

```
pfl_subset = pfl_subset.query("""
    NVS_Training_Owner_Status == 'Exists' & \
    NVS_Course_Regulatory_Classification.isna() """)
print(pfl_subset.shape)
```

```
pfl_subset = pfl_subset.query("""
    NVS_Course_Training_Title.str.startswith('[')==False
    """)
print(pfl_subset.shape)
```

```
pfl_subset = pfl_subset.query("""
    NVS_Training_Action == 'Keep'
    """)
print(pfl_subset.shape)
```

```
pfl_subset = pfl_subset.query("""
    NVS_Course_Training_Language == 'English (US)'
    """)
print(pfl_subset.shape)
```

```
pfl_subset = pfl_subset[['NVS_Training_Object_ID','NVS_Training_Action', 'NVS_Course_Training_Provider', 'NVS_Course_Training_Type', 'NVS_Course_Training_Title','NVS_Course_Description']]
pfl_subset
```

```
pfl_subset.to_csv('clean_subset.csv',index=False)
```
